```python
import pandas as pds

sourceDatePandasDateTimeIndex=pds.date_range(start='1/1/2023', end='31/12/2025')

sourceDateSparkDF=spark.createDataFrame(sourceDatePandasDateTimeIndex.to_frame(name='calendar_date'))

display(sourceDateSparkDF)
```

```python
sourceDateSparkDF.createOrReplaceTempView("source_date_dim")
```

```python
%sql
select * from source_date_dim
```

```python
%sql
truncate table `pricing-analytics`.gold.reporting_dim_date_gold;
insert into `pricing-analytics`.gold.reporting_dim_date_gold
select 
substr(calendar_date,1,10)
,date_format(calendar_date,'yyyyMMdd') 
,current_timestamp
,current_timestamp
from source_date_dim;
```
