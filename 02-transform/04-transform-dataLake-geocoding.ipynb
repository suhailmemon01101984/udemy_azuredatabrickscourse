"

```python
geoLocationSourceLayerName = 'bronze-dev'
geoLocationSourceStorageAccountName = 'dbrkcrse20251storagedev'
geoLocationSourceFolderName = 'geo-location'

geoLocationSourceFolderPath = f"abfss://{geoLocationSourceLayerName}@{geoLocationSourceStorageAccountName}.dfs.core.windows.net/{geoLocationSourceFolderName}"
```

```python
geoLocationBronzeDF = (spark
                       .read
                       .json(geoLocationSourceFolderPath))
```

```python
display(geoLocationBronzeDF)
```

```python
geoLocationSilverTransDF=geoLocationBronzeDF.select("admin1", "admin2", "country", "latitude", "longitude", "name", "population").withColumnRenamed('admin1', 'stateName').withColumnRenamed('admin2', 'districtName').withColumnRenamed('country', 'countryName').withColumnRenamed('name', 'marketName') 
```

```python
geoLocationSilverTransDF.display()
```

```python
(geoLocationSilverTransDF 
.write
.mode('overwrite')
.saveAsTable("`pricing-analytics`.silver.geo_location_silver"))
```
